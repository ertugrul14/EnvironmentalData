* **Where is the project situated?**

The project is situated in Nettelnburg, Hamburg, Germany.

*   **What is the research aim?**

This project investigates the transformation of agricultural land into urban areas in Nettelnburg, Hamburg, and its correlation with economic development patterns. The aim is to evaluate how shifts in land use have influenced economic indicators such as property prices and local infrastructure growth.

*   **What methods did you use?**

Satellite Image Retrieval
Fetched Sentinel-2 images from Google Earth Engine for each year (2018–2023), selecting the least cloudy image in June.

Image Preprocessing and Segmentation
Processed images using OpenCV and HSV color space to segment green areas. Applied morphological operations for accuracy.

Quantitative Analysis
Calculated green area percentages and converted them to hectares. Computed year-over-year changes.

Statistical Correlation
Analyzed the correlation between green area change and GDP growth using Pearson correlation and regression plots.

*   **What KPI's did you extract and why?**

Green Area Percentage
Indicates the extent of vegetation and potential urban cooling and biodiversity benefits.

Green Area in Hectares
Allows standardized comparison over the 6.25 km² area.

Year-over-Year Green Area Change (%)
Captures temporal trends in urban greening efforts or degradation.

GDP Growth (%)
Reflects economic development, included to explore the socio-ecological relationship.

Correlation Coefficient (Green Area Change vs. GDP Growth)
Evaluates whether economic growth aligns with or diverges from ecological performance.

#### **Project Results**
*(bullet points with key outcomes and learnings)*
*   Outcomes

Green areas fluctuated over the years, with some increase toward 2021–2022.

The strongest green coverage was observed in years with relatively lower cloud cover.

Calculated Pearson correlation: 0.30, suggesting a weak positive correlation between GDP growth and green area expansion.

Clear visualizations of segmented green areas helped confirm the robustness of segmentation methods.

*   Limitations

Sentinel-2 imagery at 10m resolution lacks fine-scale detail for small green patches.

Cloud coverage still influenced image availability/quality, despite filtering.

GDP data is at city level (Hamburg) rather than specific to Nettelnburg.

*   Future Work

Apply deep learning segmentation (e.g., U-Net or SAM) for more precise land classification.

Incorporate socio-demographic or policy data to contextualize environmental change.

Expand to multi-city analysis for benchmarking.
